```python
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('apeach.png')

#그레이스 케일로 변환
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#역치 지정하기(흰색과 검정색으로 구분할 것, 90 이상: 흰색, 90 이하: 검은색 )
th = 90 #기준을 90으로 할것
img[img > th] = 255 #흰색으로
img[img < th] = 0  #검은색으로 

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('apeach.png')

#이미지 엣지 검출
img = cv2.cvtColor(img, cv2.COLOR_BGR2BGRA)
img = cv2.Canny(img, 100, 200, 255)
img = 255 -img

#이미지 저장
cv2.imwrite('apeach.png', img)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

# 반전 처리
 좌우 반전 --> 1
 상하 반전 --> 0 

```python
#원본사진 -> 반전 처리 할 것 (상하반전)
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('ape.png')

#원본을 좌측에 배열
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

#이미지 상하 반전(0)
plt.subplot(1,2, 2)
img2 = cv2.flip(img, 0) # flip = 반전 0= 상하 반전 

           

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#원본사진 -> 반전 처리 할 것 (좌우 반전 )
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('ape.png')

#원본을 좌측에 배열
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

#이미지 상하 반전(0)
plt.subplot(1,2, 2)
img2 = cv2.flip(img, 1) # flip = 반전 , 1 = 좌우 반전 

           

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# 캐스케이드파일 지정
cascade_file = "haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)

# 이미지를 호출한 후 그레이 스케일로 변경
img = cv2.imread("mmmm.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 얼굴 인식하기
face_list = cascade.detectMultiScale(img_gray, minSize=(50,50))

# 결과 확인
if len(face_list) == 0 :
    print("얼굴인식 실패")
    quit()
    
# 인식한 부분 출력
for (x,y,w,h) in face_list :
    print("얼굴좌표 :", x, y,w,h)
    red = (0, 0, 255)
    cv2.rectangle(img, (x,y), (x+w, y+h), red, thickness=10)
    
    

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('on') # 이미지 격자 출력 ON/OFF
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# 캐스케이드파일 지정
cascade_file = "haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file)

# 이미지를 호출한 후 그레이 스케일로 변경
img = cv2.imread("shin.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 얼굴 인식하기
face_list = cascade.detectMultiScale(img_gray, minSize=(80,80))

# 결과 확인
if len(face_list) == 0 :
    print("얼굴인식 실패")
    quit()
    
# 인식한 부분 출력
for (x,y,w,h) in face_list :
    print("얼굴좌표 :", x, y,w,h)
    mint= (242, 236, 47)
    cv2.rectangle(img, (x,y), (x+w, y+h), mint, thickness=10)
    img2 = img[x:x+w, y:y+h]
    img2 = cv2.resize(img2, (400, 400))
    

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.axis('on') # 이미지 격자 출력 ON/OFF
plt.show()
```
